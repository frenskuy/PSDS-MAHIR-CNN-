```python
import numpy as np

def conv2d_single_channel(X, K, stride=1, padding=0):
    """
    X : array 2D (H, W)        -> input
    K : array 2D (k_h, k_w)    -> kernel/filter
    stride : int               -> langkah pergeseran kernel
    padding : int              -> jumlah padding nol di tiap sisi

    return: array 2D (H_out, W_out)
    """
    # Tambahkan padding nol di sekeliling citra
    if padding > 0:
        X_padded = np.pad(
            X,
            pad_width=((padding, padding), (padding, padding)),
            mode="constant",
            constant_values=0
        )
    else:
        X_padded = X

    H, W = X_padded.shape
    k_h, k_w = K.shape

    # Hitung ukuran output sesuai rumus
    H_out = (H - k_h) // stride + 1
    W_out = (W - k_w) // stride + 1

    Y = np.zeros((H_out, W_out), dtype=float)

    # Lakukan operasi "convolution" (lebih tepatnya cross-correlation)
    for i in range(H_out):
        for j in range(W_out):
            # koordinat awal patch pada X_padded
            i_start = i * stride
            j_start = j * stride

            # ambil patch dengan ukuran sama seperti kernel
            patch = X_padded[i_start:i_start+k_h, j_start:j_start+k_w]

            # operasi dot: sum_{m,n} patch(m,n) * K(m,n)
            Y[i, j] = np.sum(patch * K)

    return Y
```
